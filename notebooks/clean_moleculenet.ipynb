```python
from rdkit import Chem
def canonicalize_smiles(smiles: str):
    """
    Canonicalize a SMILES using rdkit
    """
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        print('Cannot canonicalize SMILES:', smiles)
        return None
    return Chem.MolToSmiles(mol, canonical=True)
```

```python
dd = pd.read_csv(os.path.join(out_dir, "esol.csv"))
print(len(dd))
dd = dd.drop_duplicates().reset_index(drop=True)
print(len(dd))
dd['canonical_smiles'] = dd['smiles'].apply(lambda x: canonicalize_smiles(x))
dd
```

```python
dd[dd["canonical_smiles"].duplicated(keep=False)].sort_values("canonical_smiles")
```

```python
dd.iloc[680,:]
```

```python
dd.iloc[1069,:]
```

```python
dd.iloc[680,0] == dd.iloc[1069,0]
```

```python
col = "measured log solubility in mols per litre"

dd_avg = (
    dd
    .groupby("canonical_smiles", as_index=False)[col]
    .mean()
)
```

```python
import pandas as pd

df_test = pd.DataFrame({
    "smiles": [
        "C(C)O",
        "CCO",
        "CC(=O)O",
        "CC(=O)O",
        "BrC1=CC=CC=C1",
        "c1ccc(Br)cc1"
    ],
    "canonical_smiles": [
        "CCO",              # duplicate
        "CCO",              # duplicate
        "CC(=O)O",          # duplicate
        "CC(=O)O",          # duplicate
        "Brc1ccccc1",       # duplicate
        "Brc1ccccc1"        # duplicate
    ],
    "measured_log_sol": [
        -0.8,
        -1.0,
        -2.1,
        -2.3,
        -4.4,
        -4.6
    ]
})

df_test
```

```python
col = "measured_log_sol"

dd_avg = (
    df_test
    .groupby("canonical_smiles", as_index=False)[col]
    .mean()
)

dd_avg
```
